```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_opioid=pd.read_csv('C:/Users/17207/CTBA/M2/OpidsVA.csv')
df_pres=pd.read_csv('C:/Users/17207/CTBA/M2/presidents.csv')
df_vehicle=pd.read_csv('C:/Users/17207/CTBA/M2/TOTALNSA.csv')
type(df_vehicle)
# Imported pandas and matplotlib packages. Pandas was also used to import the .csv files, which were placed into 
# DataFrame objects.
```

Do opioid overdoes tend to be associated with less affluent areasâ€”that is, areas where families have lower incomes?

```python
# Uncomment to verify data
# print(df_opioid)

# When reviewing the data, the best choice of graph was the scatterplot because it is used to illustrate a relationship or 
# correlation between two variables.
fig,ax = plt.subplots()
ax.scatter(df_opioid['FPOO-Rate'],df_opioid['MedianHouseholdIncome'])
ax.xaxis.set_label_text('Opioid Overdoses Per 100,000 County Residents', fontsize = 14)
ax.yaxis.set_label_text('Median Household Income', fontsize = 14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

fig.suptitle('Relationship Between Opioid Overdoses and Wealth in VA Counties', fontsize = 20)
fig.set_size_inches(10,7)

plt.show()
```

Some people who start with opioid addictions are reported to transition to heroin use. What is the relationship in Virginia counties between opioid overdoses and heroin overdoses?

```python
# Once again the scatterplot is the best choice to graph this data because it is the ideal way to plot the relationship
# between two variables.
fig,ax = plt.subplots()
ax.scatter(df_opioid['FPOO-Rate'],df_opioid['FFHO-Rate'])
ax.xaxis.set_label_text('Opioid Overdoses Per 100,000 County Residents', fontsize = 14)
ax.yaxis.set_label_text('Fentanyl/Heroin Overdoses per 100,000 County Residents', fontsize = 14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

fig.suptitle('Relationship Between Opioid Overdoses and Heroin/Fentanyl Overdoses in VA Counties', fontsize = 20)
fig.set_size_inches(10,7)

plt.show()
```

Which states are associated with the greatest number of United States presidents in terms of the presidentsâ€™ birthplaces?

```python
# In the case of this graph, a histogram is the best choice because it allows the reader to observe frequency of occurrence.
state_freq = dict(df_pres['State'].value_counts())
x = list(state_freq.keys())
height = list(state_freq.values())

fig, ax = plt.subplots()
ax.bar(x=x, height=height)
ax.xaxis.set_label_text('State', fontsize = 14)
ax.yaxis.set_label_text('Frequency', fontsize = 14)
ax.xaxis.set_tick_params(labelsize=12, rotation = 90)
ax.yaxis.set_tick_params(labelsize=12)

fig.suptitle('Frequency of US Presidents\' Birthplace', fontsize = 20)
fig.set_size_inches(15,9)

plt.show()
```

How have vehicle sales in the United States varied over time?

```python
# Uncomment to verify data
# print(df_vehicle)

# For time-series data the best way to plot the data is using a line graph, as this is the ideal graph to show changes over
# time.
x = list(df_vehicle['DATE'])
y = list(df_vehicle['TOTALNSA'])

fig, ax = plt.subplots()
ax.plot(x, y)
ax.xaxis.set_label_text('Date', fontsize = 14)
ax.yaxis.set_label_text('Total NSA', fontsize = 14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

fig.suptitle('United States Vehicle Sales Over Time', fontsize = 20)
fig.set_size_inches(15,9)

plt.show()
```
