```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_constTSdata=pd.read_csv('./ConstructionTimeSeriesDataV2.csv')
```

```python
x = list(df_constTSdata['Month'])
y = list(df_constTSdata['Private Construction'])
y1 = list(df_constTSdata['Public Construction'])

fig, ax = plt.subplots()
ax.plot(x, y, label = 'Private Construction', color = 'r')
ax.plot(x, y1, label = 'Public Construction', color = 'b')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_ylim(0,max(y)+5)
ax.set_xlabel('Day',fontsize='16')
ax.set_ylabel('Construction Costs',fontsize='16')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

ax.legend(loc=1, fontsize='12')
ax.axvspan(80,120,ymin=0,ymax = 300,facecolor='k',alpha=0.1)
ax.text(100,65000,'Private\nConstruction\nCosts\nDecreasing',fontsize=12,verticalalignment='center',horizontalalignment='center')

fig.set_size_inches(12,7)
fig.suptitle("Time Series Plot of Construction Costs (Public and Private)",fontsize='22')
```

The .csv data indicates a sharp decrease in private construction spending beginning at roughly day 80, which finally begin to rise again at about day 120.  This sharp decrease is worth noting, and is pointed out by greyscale shading around those days as well as text calling out the decrease which has been embedded into the graph.
